```python
import pandas as pd
df = pd.read_csv('antibiotic_resistance_dataset.csv') 
df.head()
```

```python
X = df.drop('resistant', axis=1)   
y = df['resistant']              
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
print("Training set size:", len(X_train))
print("Test set size:", len(X_test))
```

```python
from sklearn.preprocessing import LabelEncoder
encoders = {}
for col in ['gender', 'specimen', 'organism', 'antibiotic']:
    le = LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col])
    X_test[col] = le.transform(X_test[col])    
    encoders[col] = le
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc:.2f} ({acc*100:.1f}%)")
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, target_names=['Sensitive', 'Resistant']))
```

```python
import joblib
joblib.dump(model, 'prediction_model.pkl')
joblib.dump(encoders, 'feature_encoders.pkl')
print("prediction_model and feature_encoders saved.")
```
