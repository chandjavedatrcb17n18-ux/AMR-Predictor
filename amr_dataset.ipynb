```python
import pandas as pd
import numpy as np
np.random.seed(42)
n_samples = 500
ages = np.random.randint(1, 91, size=n_samples)      
genders = np.random.choice(['M', 'F'], size=n_samples)
specimens = np.random.choice(['urine', 'blood', 'sputum'], size=n_samples)
organisms = np.random.choice(
    ['E. coli', 'K. pneumoniae', 'S. aureus', 'P. aeruginosa'],
    size=n_samples
)
prob = resistance_rates[org][ab_idx]
antibiotics = np.random.choice(
    ['Ciprofloxacin', 'Ceftriaxone', 'Meropenem', 'Levofloxacin', 'Amikacin'],
    size=n_samples
)
resistance_rates = {
    'E. coli':         [0.65, 0.60, 0.10, 0.55, 0.15],
    'K. pneumoniae':   [0.60, 0.65, 0.15, 0.50, 0.20],
    'S. aureus':       [0.45, 0.40, 0.05, 0.40, 0.10],
    'P. aeruginosa':   [0.40, 0.50, 0.20, 0.45, 0.15]
}
antibiotic_list = ['Ciprofloxacin', 'Ceftriaxone', 'Meropenem', 'Levofloxacin', 'Amikacin']
resistant = np.zeros(n_samples, dtype=int)
for i in range(n_samples):
    org = organisms[i]
    ab = antibiotics[i]
    ab_idx = antibiotic_list.index(ab)
    resistant[i] = np.random.choice([1, 0], p=[prob, 1-prob])
df = pd.DataFrame({
    'age': ages,
    'gender': genders,
    'specimen': specimens,
    'organism': organisms,
    'antibiotic': antibiotics,
    'resistant': resistant
})
print("First 10 rows:")
print(df.head(10))
print("\nDataset shape:", df.shape)
df.to_csv('amr_dataset.csv', index=False)
print("\nData saved as 'amr_dataset.csv'")
```

```python

print("Overall resistance rate:", df['resistant'].mean()*100, "%")
print("\nResistance by organism:")
print(df.groupby('organism')['resistant'].mean())
print("\nResistance by antibiotic:")
print(df.groupby('antibiotic')['resistant'].mean())
```
